# list

- lists are a fundamental data structure used to store collections of items in an ordered and changeable way.
- You can create lists using square brackets `[]` and separate items with commas.
- Lists can contain items of different data types (strings, integers, floats, booleans, even other lists) within the same list.
- Elements in a list are indexed starting from 0. The first element has index 0, the second element has index 1, and so on.
- You can access elements using their index within square brackets.
- Negative indexing allows you to access elements from the end of the list. -1 refers to the last element, -2 to the second-last, and so on.
- You can extract a portion of a list using slicing with square brackets and colons (`:`)
- The syntax is `[start:end:step]`
    - `start` (optional): Index of the first element to include (inclusive). Defaults to 0.
    - `end` (optional): Index of the element to exclude (exclusive). Defaults to the list length.
    - `step` (optional): Number of steps to take between elements. Defaults to 1 (iterates through all elements).
- Lists are mutable, meaning you can change their contents after creation.
- You can assign a new value to an element using its index.
- methods to add or remove elements from a list
    - `append(element)`: Add an element to the end of the list.
    - `insert(index, element)`: Insert an element at a specific index.
    - `remove(element)`: Remove the first occurrence of an element (error if not found).
    - `pop(index)`: Remove and return the element at a specific index (or the last element by default).
- Common List Operations
    - `len(list)`: Returns the length (number of items) in the list.
    - `in` and `not in`: Check if an element is present in the list.
    - `list.extend(iterable)`: Extends the list by appending all elements from another iterable (like another list).
- Key Points
    - Lists are versatile and can store various data types.
    - Indexing provides access to individual elements.
    - Slicing allows you to extract sublists.
    - Lists are mutable, enabling modification of their contents.
    - Various methods offer functionalities for adding, removing, and manipulating list elements.

```python
# create list
fruits = ["ðŸŽ", "ðŸŒ", "ðŸ’"]
print(fruits)       # ['ðŸŽ', 'ðŸŒ', 'ðŸ’']

numbers = [1, 2, 3, 4.5]
print(numbers)      # [1, 2, 3, 4.5]

mixed_list = ["hello", 10, True]
print(mixed_list)   # ['hello', 10, True]
```

```python
# get items from list
first_fruit = fruits[0]  # first_fruit will be "ðŸŽ"
print(first_fruit)  # ðŸŽ

last_number = numbers[-1]  # -1 refers to the last element, last_number will be 4.5
print(last_number)  # 4.5

# appending items to list
fruits.append("ðŸ¥­") # ['ðŸŽ', 'ðŸŒ', 'ðŸ’', 'ðŸ¥­']
print(fruits)
```

```python
# slicing list

# s[i:j] ðŸ‘ˆ slice of s from i to j
print(fruits[1:3])      # ['ðŸŒ', 'ðŸ’']

# s[i:j:k] ðŸ‘ˆ slice of s from i to j with step k
print(fruits[1:4:2])    # ['ðŸŒ', 'ðŸ’']
print(fruits[::3])      # ['ðŸŽ', 'ðŸ¥­'] (elements at index 0 and 3, takes every 3rd item starting from item 0)
print(fruits[::-3])     # ['ðŸ¥­', 'ðŸŽ'] (elements at index -1 and -3, takes every 3rd item starting from item -1)

# traversing within list
for fruit in fruits:
    print(fruit)
```

```python
# copying list
fruits_copy = fruits.copy()
print(fruits_copy)  # ['ðŸŽ', 'ðŸŒ', 'ðŸ’', 'ðŸ¥­']

# replacing item in list
fruits_copy[1] = "ðŸ¥­"  # Replace "ðŸŒ" with "ðŸ¥­"
print(fruits_copy)  # ['ðŸŽ', 'ðŸ¥­', 'ðŸ’', 'ðŸ¥­']
```

```python
# list comprehension.
squares = [i**2 for i in range(10)]
print(squares)  # [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```

```python
# x in s ðŸ‘ˆ True if an item of s is equal to x, else False
print("ðŸ¥­" in fruits)       # True

# x not in s ðŸ‘ˆ False if an item of s is equal to x, else True
print("ðŸ¥­" not in fruits)   # False

# s + t ðŸ‘ˆ the concatenation of s and t
print(fruits + squares)     # ['ðŸŽ', 'ðŸ¥­', 'ðŸ’', 'ðŸ¥­', 0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# s * n or n * s ðŸ‘ˆ equivalent to adding s to itself n times
print(fruits * 2)           # ['ðŸŽ', 'ðŸ¥­', 'ðŸ’', 'ðŸ¥­', 'ðŸŽ', 'ðŸ¥­', 'ðŸ’', 'ðŸ¥­']
```

```python
# len(s) ðŸ‘ˆ length of s
print(len(squares))

# min(s) ðŸ‘ˆ smallest item of s
print(min(squares))

# max(s) ðŸ‘ˆ largest item of s
print(max(squares))

# s.index(x[, i[, j]]) ðŸ‘ˆ index of the first occurrence of x in s (at or after index i and before index j)
print(fruits_copy.index('ðŸ¥­', 2, 4))

# s.count(x) ðŸ‘ˆ total number of occurrences of x in s
print(fruits_copy.count('ðŸ¥­'))

# s.append(x) ðŸ‘ˆ appends x to the end of the sequence (same as s[len(s):len(s)] = [x])
fruits_copy.append("ðŸŠ")
print(fruits_copy)          # ['ðŸ’', 'ðŸŒ', 'ðŸŽ', 'ðŸ’', 'ðŸŒ', 'ðŸŽ', 'ðŸ¥‘', 'ðŸŠ']

# s.clear() ðŸ‘ˆ removes all items from s (same as del s[:])
fruits_copy.clear()
print(fruits_copy)          # []

# s.extend(t) or s += t ðŸ‘ˆ extends s with the contents of t (for the most part the same as s[len(s):len(s)] = t)
fruits_copy.extend(fruits)
print(fruits_copy)          # ['ðŸŽ', 'ðŸŒ', 'ðŸ’', 'ðŸ¥­']

# s *= n ðŸ‘ˆ updates s with its contents repeated n times
fruits_copy *= 2
print(fruits_copy)          # ['ðŸŽ', 'ðŸŒ', 'ðŸ’', 'ðŸ¥­', 'ðŸŽ', 'ðŸŒ', 'ðŸ’', 'ðŸ¥­']

# s.insert(i, x) ðŸ‘ˆ inserts x into s at the index given by i (same as s[i:i] = [x])
fruits_copy.insert(1, "ðŸ¥‘")
print(fruits_copy)          # ['ðŸŽ', 'ðŸ¥‘', 'ðŸŒ', 'ðŸ’', 'ðŸ¥­', 'ðŸŽ', 'ðŸŒ', 'ðŸ’', 'ðŸ¥­']

# s.pop() or s.pop(i) ðŸ‘ˆ retrieves the item at i and also removes it from s
print(fruits_copy.pop())    # ðŸ¥­

# s.remove(x) ðŸ‘ˆ remove the first item from s where s[i] is equal to x
fruits_copy.remove("ðŸ¥­")
print(fruits_copy)          # ['ðŸŽ', 'ðŸ¥‘', 'ðŸŒ', 'ðŸ’', 'ðŸŽ', 'ðŸŒ', 'ðŸ’']

# s.reverse() ðŸ‘ˆ reverses the items of s in place
fruits_copy.reverse()
print(fruits_copy)          # ['ðŸ’', 'ðŸŒ', 'ðŸŽ', 'ðŸ’', 'ðŸŒ', 'ðŸ¥‘', 'ðŸŽ']
```

```python
# sorting
from random import randint

random_numbers = [randint(0, 100) for _ in range(10)]
print(random_numbers)

random_numbers.sort()
print(random_numbers)

random_numbers.sort(reverse=True)
print(random_numbers)

random_numbers.sort(key=lambda x: x > 50, reverse=False)
print(random_numbers)
```
